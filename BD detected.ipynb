```python
import requests
from bs4 import BeautifulSoup
from github import Installation

# URL objetivo
url = "https://www.gob.pe/institucion/minem/colecciones/6-boletin-estadistico-minero"

# Headers (importante para evitar bloqueos)
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Request
response = requests.get(url, headers=headers)
response.raise_for_status()

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Contenedor principal
container = soup.find("div", class_="w-full js-official-documents-search-results")

# Encontrar todos los bloques
items = container.find_all("div", class_="col-md-6 py-2")

datos = []

for item in items:
    card = item.find("div", class_="p-4 shadow-campaign-card border-b-5 border-black h-full flex flex-col justify-between")

    # Fecha
    fecha = card.find("div", class_="border-b border-gray-700 pb-2 font-medium mb-3")
    fecha_text = fecha.get_text(strip=True) if fecha else None

    # Título y link
    a_tag = card.find("a", class_="leading-6 font-bold underline")
    titulo = a_tag.get_text(strip=True) if a_tag else None
    link = "https://www.gob.pe" + a_tag["href"] if a_tag else None

    # Descripción (texto libre dentro del card)
    descripcion = None
    textos = card.find_all(text=True, recursive=False)
    if textos:
        descripcion = " ".join(t.strip() for t in textos if t.strip())

    # Formatos
    formatos = card.find("div", class_="mt-3 leading-5 text-sm")
    formatos_text = formatos.get_text(strip=True) if formatos else None

    datos.append({
        "fecha": fecha_text,
        "titulo": titulo,
        "link": link,
        "descripcion": descripcion,
        "formatos": formatos_text
    })

# Mostrar resultados
for d in datos:
    print(d)
```

```python
import pandas as pd

# Crear DataFrame desde tu lista
df2 = pd.DataFrame(datos)

# Seleccionar solo columnas necesarias
df2 = df2[["fecha", "descripcion"]]

# Eliminar filas vacías (por seguridad)
df2 = df2.dropna()

# Resetear índice
df2 = df2.reset_index(drop=True)

# Mostrar resultado
print(df2)
```

```python
import requests
import pandas as pd
from datetime import datetime
import os

# ================= CONFIG =================
ruta_f2 = r"\\sppeapp00023\Inversiones\Mesa de Inversiones\Bottom-Up\1 Proceso de Inversión Renta Variable\Proyecto Inputs Macro\Alerta BD\Financial Alert.xlsx"

indicadores = {
    "B-2201": "EGP",
    "B-2401": "Indicadores",
    "B-2309": "Creditos"
}

meses = [
    ("di", "Diciembre"),
    ("no", "Noviembre"),
    ("oc", "Octubre"),
    ("se", "Setiembre"),
    ("ag", "Agosto"),
    ("jl", "Julio"),
    ("jn", "Junio"),
    ("my", "Mayo"),
    ("abr", "Abril"),
    ("ma", "Marzo"),
    ("fe", "Febrero"),
    ("en", "Enero"),
]

headers = {"User-Agent": "Mozilla/5.0"}


# ================= FUNCIONES =================

def existe_url(url):
    try:
        r = requests.head(url, headers=headers, timeout=5)
        return r.status_code == 200
    except:
        return False


def encontrar_ultimo(indicador):
    current_year = datetime.now().year

    for anio in range(current_year, current_year - 5, -1):
        for cod_mes, nombre_mes in meses:
            archivo = f"{indicador}-{cod_mes}{anio}.XLS"
            url = f"https://intranet2.sbs.gob.pe/estadistica/financiera/{anio}/{nombre_mes}/{archivo}"

            if existe_url(url):
                return {
                    "fecha": f"{nombre_mes} {anio}",
                    "Link": url,
                    "descripcion": f"{indicador} - {nombre_mes} {anio}",
                    "link": url
                }

    return None


# ================= 1. SCRAPING =================

data_por_hoja = {}

for indicador, nombre_hoja in indicadores.items():

    resultado = encontrar_ultimo(indicador)

    if resultado:
        df = pd.DataFrame([resultado])
    else:
        df = pd.DataFrame(columns=["fecha", "Link", "descripcion", "link"])

    data_por_hoja[nombre_hoja] = df


# ================= 2. CARGAR HISTÓRICO =================

excel_anterior = {}

if os.path.exists(ruta_f2):
    try:
        excel_anterior = pd.read_excel(ruta_f2, sheet_name=None, engine="openpyxl")
        print("📂 Histórico cargado")
    except Exception as e:
        print("⚠️ Error leyendo histórico:", e)
else:
    print("📁 Primera ejecución (no hay histórico)")


# ================= 3. DETECCIÓN DE CAMBIOS =================

lista_alertas = []

for hoja, df_nuevo in data_por_hoja.items():

    df_viejo = excel_anterior.get(hoja)

    if df_viejo is None:
        df_nuevas = df_nuevo.copy()
    else:
        columnas_comunes = list(set(df_nuevo.columns).intersection(set(df_viejo.columns)))

        df_nuevo_sub = df_nuevo[columnas_comunes].astype(str)
        df_viejo_sub = df_viejo[columnas_comunes].astype(str)

        df_nuevas = df_nuevo_sub.merge(
            df_viejo_sub,
            how="outer",
            indicator=True
        )

        df_nuevas = df_nuevas[df_nuevas["_merge"] == "left_only"].drop(columns="_merge")

    # ===== PRINT ALERTAS =====
    if not df_nuevas.empty:
        df_nuevas["fuente"] = hoja
        lista_alertas.append(df_nuevas)

        print(f"\n🚨 ALERTA en {hoja}: {len(df_nuevas)} cambio(s)")
        print(df_nuevas)

    else:
        print(f"\n✅ {hoja}: sin cambios")


# ================= 4. DECISIÓN FINAL =================

if lista_alertas:
    print("\n💾 Se detectaron cambios → sobrescribiendo Excel...")

    with pd.ExcelWriter(ruta_f2, engine="openpyxl") as writer:
        for hoja, df in data_por_hoja.items():
            df.to_excel(writer, sheet_name=hoja, index=False)

    print("✅ Archivo actualizado")

else:
    print("\n⏸ No hubo cambios → no se modifica el archivo")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
import pandas as pd
import urllib.parse
import re
import time

# Configuración
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 15)

url = "https://www.ine.gob.cl/estadisticas-por-tema/industria-energia-y-construccion/indice-de-produccion-industrial"
driver.get(url)

# Click en BOLETINES
boletines = wait.until(EC.element_to_be_clickable((
    By.CSS_SELECTOR, 'div.navPrincipalDescargas[data-folder="fc62c4c1-8e79-4a42-bfc6-b37238816f98"]'
)))
boletines.click()

# Año actual
year_actual = str(datetime.now().year)

# Click en el año
anio_elemento = wait.until(EC.element_to_be_clickable((
    By.XPATH, f'//div[contains(@class,"categoriaDescarga") and text()="{year_actual}"]'
)))
anio_elemento.click()

time.sleep(2)

# Extraer links
links = driver.find_elements(By.CSS_SELECTOR, 'a.widArchNavArchivoDescarga')
hrefs = [link.get_attribute("href") for link in links]

driver.quit()

# --- Procesamiento para dataframe ---

data = []
titulo_fijo = "Indice de Produccion Industrial"

for url in hrefs:
    # Decodificar URL
    url_decoded = urllib.parse.unquote(url.lower())

    # Extraer mes y año desde el nombre del archivo
    match = re.search(r'(enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre)-(\d{4})', url_decoded)

    if match:
        mes = match.group(1)
        anio = match.group(2)
        fecha = f"{mes} {anio}"
    else:
        fecha = None

    data.append({
        "titulo": titulo_fijo,
        "fecha": fecha
    })

# Crear DataFrame
df_ipi = pd.DataFrame(data)

print(df_ipi)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
import pandas as pd
import re
import time

# Configuración
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 15)

url = "https://www.ositran.gob.pe/anterior/publicaciones/boletin-estadistico/"
driver.get(url)

# Año actual
year_actual = str(datetime.now().year)

# 1. Click en el tab del año actual
tab_anio = wait.until(EC.element_to_be_clickable((
    By.XPATH, f'//ul[contains(@class,"nav-tabs")]//a[text()="{year_actual}"]'
)))
tab_anio.click()

# 2. Esperar a que el contenido del año cargue
wait.until(EC.presence_of_element_located((
    By.ID, f"tab-{year_actual}"
)))

time.sleep(1)  # pequeña pausa por estabilidad

# 3. Extraer boletines
elementos = driver.find_elements(
    By.CSS_SELECTOR, f'#tab-{year_actual} div.bole-item h2 a'
)

data = []

for el in elementos:
    href = el.get_attribute("href")
    titulo_raw = el.text  # "Boletín Estadístico | Abril 2026"

    # Extraer mes y año
    match = re.search(r'(\w+)\s(\d{4})', titulo_raw.lower())

    if match:
        mes = match.group(1)
        anio = match.group(2)
        fecha = f"{mes} {anio}"
    else:
        fecha = None

    data.append({
        "titulo": "Boletín Estadístico OSITRAN",
        "fecha": fecha
    })

driver.quit()

# 4. Crear DataFrame
df_ositran = pd.DataFrame(data)

print(df_ositran)
```

```python
import pandas as pd
import os

# ===== RUTA =====
ruta_construction = r"\\sppeapp00023\Inversiones\Mesa de Inversiones\Bottom-Up\1 Proceso de Inversión Renta Variable\Proyecto Inputs Macro\Alerta BD\Construction Alert.xlsx"

# ===== DATA NUEVA =====
df_ipi_nuevo = df_ipi.copy()
df_ositran_nuevo = df_ositran.copy()

# ===== LEER HISTÓRICO =====
if os.path.exists(ruta_construction):
    try:
        df_ipi_viejo = pd.read_excel(ruta_construction, sheet_name="IPI-Chile", engine="openpyxl")
        df_ositran_viejo = pd.read_excel(ruta_construction, sheet_name="OSITRAN-Peru", engine="openpyxl")
        print("📂 Histórico construction cargado")
    except Exception as e:
        print("⚠️ Error leyendo histórico:", e)
        df_ipi_viejo, df_ositran_viejo = None, None
else:
    print("📁 No existe histórico (primera ejecución)")
    df_ipi_viejo, df_ositran_viejo = None, None


# ================= DETECCIÓN DE CAMBIOS =================
nuevas_alertas_construccion = []

def detectar_nuevos(df_nuevo, df_viejo, nombre_fuente):
    if df_viejo is None:
        df_nuevas = df_nuevo.copy()
    else:
        columnas_comunes = list(set(df_nuevo.columns).intersection(set(df_viejo.columns)))

        df_nuevo_sub = df_nuevo[columnas_comunes].astype(str)
        df_viejo_sub = df_viejo[columnas_comunes].astype(str)

        df_nuevas = df_nuevo_sub.merge(
            df_viejo_sub,
            how="outer",
            indicator=True
        )

        df_nuevas = df_nuevas[df_nuevas["_merge"] == "left_only"].drop(columns="_merge")

    if not df_nuevas.empty:
        df_nuevas["fuente"] = nombre_fuente
        nuevas_alertas_construccion.append(df_nuevas)

        print(f"\n🚨 ALERTA {nombre_fuente}: {len(df_nuevas)} nuevo(s)")
        print(df_nuevas)
    else:
        print(f"\n✅ {nombre_fuente}: sin cambios")


# Ejecutar para ambas fuentes
detectar_nuevos(df_ipi_nuevo, df_ipi_viejo, "IPI Chile")
detectar_nuevos(df_ositran_nuevo, df_ositran_viejo, "OSITRAN Peru")


# ================= DECISIÓN FINAL =================
if nuevas_alertas_construccion:
    print("\n💾 Se detectaron cambios → sobrescribiendo archivo construction...")

    with pd.ExcelWriter(ruta_construction, engine='openpyxl') as writer:
        df_ipi_nuevo.to_excel(writer, sheet_name="IPI-Chile", index=False)
        df_ositran_nuevo.to_excel(writer, sheet_name="OSITRAN-Peru", index=False)

    print("✅ Archivo actualizado")

else:
    print("\n⏸ No hubo cambios → no se modifica el archivo")


# ===== CONSOLIDADO FINAL =====
if nuevas_alertas_construccion:
    nuevas_alertas_construccion = pd.concat(nuevas_alertas_construccion, ignore_index=True)
else:
    nuevas_alertas_construccion = pd.DataFrame()

print("\n📊 DataFrame final de alertas:")
print(nuevas_alertas_construccion)
```
